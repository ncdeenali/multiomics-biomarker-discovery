```python
import pandas as pd
from mofapy2.run.entry_point import entry_point
```

```python
# initializing the entry point
ent = entry_point()
```

```python
rna_df = pd.read_csv('rna_csv.csv', index_col = 0).T
rna_df.head()
```

```python
mut_df = pd.read_csv('mut_csv.csv', index_col = 0).T
mut_df.head()
```

```python
ent.set_data_options(scale_views = False)
```

```python
rna_matrix = rna_df.to_numpy()
mut_matrix = mut_df.to_numpy()
```

```python
rna_matrix.shape
```

```python
mut_matrix.shape
```

```python
ent.set_data_matrix(
    data = [[rna_matrix.T], [mut_matrix.T]],
    likelihoods = ['gaussian', 'bernoulli'])
```

```python
ent.set_model_options(
    factors = 10,
    spikeslab_weights = True,
    ard_weights = True
)
```

```python
ent.set_train_options(
    convergence_mode = "fast",
    dropR2 = 0.001,
    gpu_mode = False,
    seed = 1
)
```

```python
ent.build()
ent.run()

# Save the output
ent.save('bioinfo_mofa_model.hdf5')
```
